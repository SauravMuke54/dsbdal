```python
#here name is null so illogical to use mode so dropping is better
df.dropna(subset=['Name'],inplace=True)
```

```python
df.isna().sum()
```

```python
df
```

```python
#college is string so filling it with mode as mean and median cant be calculated on string
df['College'].fillna(df['College'].mode()[0],inplace=True) 
```

```python
df.isna().sum()
```

```python
#salary is numeric column hence fill it by mean()
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
#all null values filled data is consistent now
```

```python
#converting the data ie categorical to numerical
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['Team']=encoder.fit_transform(df['Team'])
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
cols=df['Position'].unique()
pos=pd.DataFrame(encoder.fit_transform(df[['Position']]).toarray(),columns=cols)
```

```python
pos
```

```python
df=pd.concat([df,pos],axis=1)
```

```python
df
```

```python
#min max scaler on salary
min1=df['Salary'].min()
max1=df['Salary'].max()
for i,value in enumerate(df['Salary']):
    df['Salary'][i]=((value-min1)/(max1-min1))*(1-0)+0
```
