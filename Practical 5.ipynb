```python
#now to convert quantitative into categorical by binning method
```

```python
df['Age'].value_counts()
```

```python
#creating a bin 
bins=[19,25,30,35,40]
labels=['19-24','25-29','30-34','35-40'] # creating labels for bins
df['AgeGroup']=pd.cut(df['Age'],bins=bins,labels=labels,right=False)
```

```python
df
```

```python
agegroup=df.groupby(df['AgeGroup'])
```

```python
agegroup['Salary'].describe()
```

```python
#creating a list of salary by age group
salarylist=list(agegroup['Salary'])
```
