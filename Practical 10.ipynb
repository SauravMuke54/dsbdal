```python
#1. Extract Sample document and apply following document preprocessing methods: Tokenization, POS Tagging, stop words removal, Stemming and Lemmatization.

#2. Create representation of documents by calculating Term Frequency and Inverse Document Frequency.
```

```python
import docx2txt #to convert word doc to txt file
```

```python
path ="testdoc.docx" #filename 
text1=docx2txt.process(path)
```

```python
print(text1)
```

```python
path='testdoc2.docx'
text2=docx2txt.process(path)
```

```python
print(text2)
```

```python
#tokenization of sentence 
from nltk.tokenize import sent_tokenize
sent1=sent_tokenize(text1)
sent2=sent_tokenize(text2)
print(sent1,sent2)
```

```python
#word tokenization
from nltk.tokenize import word_tokenize
word1=word_tokenize(text1)
word2=word_tokenize(text2)
print(word1,word2)
```

```python
#to check the frequency distribution of the word we use
```

```python
from nltk.probability import FreqDist
freq1=FreqDist(word1)
print(freq1)
freq2=FreqDist(word2)
print(freq2)
```

```python
#checking the frequency dist
freq1.most_common()
```

```python
freq2.most_common()
```

```python
import matplotlib.pyplot as plt
freq1.plot(20) #show number of words
plt.show()
```

```python
#now removing the stop words
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
new_word1=[] #new word list for words without stopwords
for it in word1:
    if it not in stop_words:
        new_word1.append(it)
print(new_word1)
```

```python
new_word2=[]
for it in word2:
    if it not in stop_words:
        new_word2.append(it)
print(new_word2)
```

```python
#stemming
#converting the words to its base form
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
stemmed_word1=[]
for it in new_word1:
    stemmed_word1.append(ps.stem(it))
print("Old Words")
print(new_word1)
print("Stemmed Word")
print(stemmed_word1)
```

```python
stemmed_word2=[]
for it in new_word2:
    stemmed_word2.append(ps.stem(it))
print("Old Words")
print(new_word2)
print("New Word")
print(stemmed_word2)
```

```python
#Lemmatization
#convert it to the base form by knowing its meaning 
# like better will be converted to good
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
wnl=WordNetLemmatizer()
```

```python

lemmatize_word1=[]
for it in new_word1:
    lemmatize_word1.append(wnl.lemmatize(it))
print("Old World")
print(new_word1)
print("Lemmatized Word")
print(lemmatize_word1)
```

```python
#now calculating TF term frequency
documents=[text1,text2]
for it in documents:
    dict={}
    for word in it.split():
        if word in dict:
            dict[word]+=1
        else:
            dict[word]=1
    for key,value in dict.items():
        dict[key]=value/len(it)
        print(key,dict[key])
```
